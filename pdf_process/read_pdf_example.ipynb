```python
# 引入 package
# 詳細用法可看這裡 https://github.com/chrismattmann/tika-python
from tika import parser
from tika import unpack
from tika import detector
from tika import language
from tika import translate
from pprint import pprint
```

[tika 詳細用法](https://github.com/chrismattmann/tika-python)

```python
# 下載檔案並指定檔案位置
!mkdir data
!wget wget https://www.aclweb.org/anthology/N12-1003.pdf -O data/N12.pdf
file = "data/N12-1003.pdf"
```

```python
# 讀取檔案
# 第一次跑會需要一段時間安裝需要的元件
parsed = parser.from_file(file)
```

```python
# 簡單處理後印出前 1k 個字元
parsed['content'] = parsed['content'].replace("\n",'')
pprint(parsed['content'][:1000])
```

```python
# 印出 metadata
pprint(parsed['metadata'])

```

```python
# 讀取文件後轉成 xml 格式
parsed = parser.from_file(file, xmlContent=True)
```

```python
# 印出來
pprint(parsed['content'][:10000])
```
